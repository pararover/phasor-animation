### Resistive AC circuit

Source AC voltage is:
\begin{align*}
    v(t) = v_0 \sin{\omega t}
\end{align*}

Current through the resistor is:
\begin{align*}
    i(t) = i_0 \sin{\omega t}
\end{align*}

where the peak current is
\begin{align*}
    i_0 = \dfrac{v_0}{R}
\end{align*}

Therefore the phase difference between $v(t)$ and $i(t)$ is $\phi=0$.

```python
phi = 0.0
theta_1=np.pi/4
ac_case = 'resistive'
```

```python
%matplotlib agg

frames = []

theta_array = np.linspace(0, 2*np.pi, 60)
for i, theta_1 in enumerate(theta_array):
    fig, ax = plot_phasor(phi=phi, theta_1=theta_1, v_0=0.6, i_0=1.0)
    buf = io.BytesIO()
    fig.savefig(buf, format="png", dpi=150, transparent=False)
    buf.seek(0)
    frames.append(Image.open(buf))

anim_file = 'gifs/phasor-'+ac_case+'-animation.gif'
frames[0].save(
    anim_file,
    save_all=True,
    append_images=frames[1:],
    duration=200,
    loop=0,
)

animdisp(open(anim_file,'rb').read())
```

```python
del buf
fig.clear()
plt.close()
```

### Inductive AC circuit

Source AC voltage is:
\begin{align*}
    v(t) = v_0 \sin{\omega t}
\end{align*}

Current through the inductor is:
\begin{align*}
    i(t) = i_0 \sin{\left( \omega t - \dfrac{\pi}{2} \right)}
\end{align*}

where the peak current is
\begin{align*}
    i_0 = \dfrac{v_0}{X_L}
\end{align*}

with the inductive reactance being
\begin{align*}
    X_L = \omega L
\end{align*}

Therefore the phase difference between $v(t)$ and $i(t)$ is $\phi=-\dfrac{\pi}{2}$.

```python
phi = -np.pi/2
theta_1=np.pi/4
ac_case = 'inductive'
```

```python
%matplotlib agg

frames = []

theta_array = np.linspace(0, 2*np.pi, 60)
for i, theta_1 in enumerate(theta_array):
    fig, ax = plot_phasor(phi=phi, theta_1=theta_1, v_0=0.6, i_0=1.0)
    buf = io.BytesIO()
    fig.savefig(buf, format="png", dpi=150, transparent=False)
    buf.seek(0)
    frames.append(Image.open(buf))

anim_file = 'gifs/phasor-'+ac_case+'-animation.gif'
frames[0].save(
    anim_file,
    save_all=True,
    append_images=frames[1:],
    duration=200,
    loop=0,
)

animdisp(open(anim_file,'rb').read())
```

```python
del buf
fig.clear()
plt.close()
```

### Capacitive AC circuit

Source AC voltage is:
\begin{align*}
    v(t) = v_0 \sin{\omega t}
\end{align*}

Current through the capacitor is:
\begin{align*}
    i(t) = i_0 \sin{\left( \omega t + \dfrac{\pi}{2} \right)}
\end{align*}

where the peak current is
\begin{align*}
    i_0 = \dfrac{v_0}{X_C}
\end{align*}

with the capacitive reactance being
\begin{align*}
    X_C = \dfrac{1}{\omega C}
\end{align*}

Therefore the phase difference between $v(t)$ and $i(t)$ is $\phi=+\dfrac{\pi}{2}$.

```python
phi = np.pi/2
theta_1=np.pi/4
ac_case = 'capacitive'
```

```python
%matplotlib agg

frames = []

theta_array = np.linspace(0, 2*np.pi, 60)
for i, theta_1 in enumerate(theta_array):
    fig, ax = plot_phasor(phi=phi, theta_1=theta_1, v_0=0.6, i_0=1.0)
    buf = io.BytesIO()
    fig.savefig(buf, format="png", dpi=150, transparent=False)
    buf.seek(0)
    frames.append(Image.open(buf))

anim_file = 'gifs/phasor-'+ac_case+'-animation.gif'
frames[0].save(
    anim_file,
    save_all=True,
    append_images=frames[1:],
    duration=200,
    loop=0,
)

animdisp(open(anim_file,'rb').read())
```
